```python
import pickle as pkl
import numpy as np
```

```python
from Monte_Carlo_Shapley import Monte_Carlo_Shapley
from Monte_Carlo_Asymmetric_Shapley import Monte_Carlo_Asymmetric_Shapley

# Load traffic data (split 6)

split = 6

m1, m2, m3, m4, m5, m6 = 'jul', 'aug', 'sep', 'oct', 'nov', 'dec'

with open("data/traffic/traffic_volume_split{}.pickle".format(split), "rb") as fp:
    mydict = pkl.load(fp)

X_m1_trn = mydict['X_{}_trn'.format(m1)]
y_m1_trn = mydict['y_{}_trn'.format(m1)]
X_m2_trn = mydict['X_{}_trn'.format(m2)]
y_m2_trn = mydict['y_{}_trn'.format(m2)]
X_m3_trn = mydict['X_{}_trn'.format(m3)]
y_m3_trn = mydict['y_{}_trn'.format(m3)]

X_m1m2_trn = mydict['X_{}_{}_trn'.format(m1, m2)]
y_m1m2_trn = mydict['y_{}_{}_trn'.format(m1, m2)]
X_m1m2_val = mydict['X_{}_{}_val'.format(m1, m2)]
y_m1m2_val = mydict['y_{}_{}_val'.format(m1, m2)]

X_m1m2m3_trn = mydict['X_{}_{}_{}_trn'.format(m1, m2, m3)]
y_m1m2m3_trn = mydict['y_{}_{}_{}_trn'.format(m1, m2, m3)]
X_m1m2m3_val = mydict['X_{}_{}_{}_val'.format(m1, m2, m3)]
y_m1m2m3_val = mydict['y_{}_{}_{}_val'.format(m1, m2, m3)]

X_m3 = mydict['X_{}'.format(m3)]
y_m3 = mydict['y_{}'.format(m3)]
X_m4 = mydict['X_{}'.format(m4)]
y_m4 = mydict['y_{}'.format(m4)]
X_m5 = mydict['X_{}'.format(m5)]
y_m5 = mydict['y_{}'.format(m5)]
X_m6 = mydict['X_{}'.format(m6)]
y_m6 = mydict['y_{}'.format(m6)]

s_class_dic = {0: np.asarray(range(len(X_m1_trn))),
               1: np.asarray(range(len(X_m1_trn), len(X_m1m2_trn)))}
```

```python
# Compute the symmetric Monte Carlo Shapley

seed = split # split 6, use seed 6
print('split:', split)
model = 'KNN'
problem = 'classification'
metric = 'accuracy'
K = 5

directory = 'experiments/sequential data valuation/traffic/split{}/experiment_1/'.format(split)

mc_shapley = Monte_Carlo_Shapley(X_m1m2_trn, y_m1m2_trn, X_m1m2_val, y_m1m2_val, seed=seed,
                                 directory=directory, continued=False, metric=metric,
                                 n_neighbors=K, task=problem, model_family=model,
                                 max_permutations=5000)

mc_shapley.run(50, err=0.01, tolerance=0, loo_run=True)
```

```python
# Compute the asymmetric Monte Carlo Shapley

seed = split # split 6, use seed 6
print('split:', split)
model = 'KNN'
problem = 'classification'
metric = 'accuracy'
K = 5

directory = 'experiments/sequential data valuation/traffic/split{}/experiment_1/'.format(split)

mc_asym_shapley = Monte_Carlo_Asymmetric_Shapley(X_m1m2_trn, y_m1m2_trn, X_m1m2_val, y_m1m2_val, seed=seed,
                                                 directory=directory,  social_class=s_class_dic,
                                                 continued=False, metric=metric,
                                                 n_neighbors=K, task=problem, model_family=model,
                                                 max_permutations=5000)

mc_asym_shapley.run(50, err=0.01, tolerance=0, loo_run=False)
```
